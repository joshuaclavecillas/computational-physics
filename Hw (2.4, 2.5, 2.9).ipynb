```python
#But after some research I found in openstax the equation for relativistic time dilation which is 
#delta t (stationary) = [delta tau (time of ship)]*[lorentz factor]
import numpy as np

x = 10 #no. light years away from another planet
v_c = 0.99 #velocity as a fraction of the speed of light
time_at_earth = x / v_c
#lorentz factor
lrfac = 1/np.sqrt(1 - (v_c**2)) 

time_near_c = time_at_earth / lrfac
    

    
print("If the observer is on earth the spaceship will take:", time_at_earth)
print("If the observer is on the spaceship, they will percieive:", time_near_c)
print(lrfac)
```

```python
#problem 2.5
import numpy as np
m = 9.11*10**(-31)         #mass of an electron in kg
E = 10                     #kinetic energy in eV
V = 9                      #potential step height
hbar =  1.05*10**(-34)   #reduced planck constant 
#eV_j =  1.60*10**(-19)     #electron volts to joules
#initial wave vector
k1 = np.sqrt(2*m*E)/hbar
#smaller wave vector
k2 = np.sqrt(2*m*(E-V))/hbar

#probability of transmission
T = (4*k1*k2)/(k1+k2)**2

#probability of reflection
R = ((k1 - k2)/(k1 + k2))**2

print("A particle with mass of an electron and initial kinetic energy of", E, "eV, and an potential step of", V, "eV")
print("Has a ", T*100, "% chance to transmit through and a ", R*100, "% chance to reflect")
```
