# Sierpiński Gasket — Starter Notebook

This notebook starts with the **first chaos-game step**:
1. Draw an equilateral triangle with vertices \(a_1,b_1), (a_2,b_2), (a_3,b_3)\.
2. Place an arbitrary point \(x_0,y_0)\ inside the triangle.
3. Randomly choose integer 1, 2, or 3 and place the next point halfway to the corresponding vertex.

```python
import numpy as np
import matplotlib.pyplot as plt

# Equilateral triangle vertices
a1, b1 = 0.0, 0.0
a2, b2 = 1.0, 0.0
a3, b3 = 0.5, np.sqrt(3) / 2

V1 = np.array([a1, b1])
V2 = np.array([a2, b2])
V3 = np.array([a3, b3])
vertices = [V1, V2, V3]
```

```python
# Place an arbitrary point P=(x0, y0) inside the triangle
seed = 6
rng = np.random.default_rng(seed)

# Random barycentric coordinates guarantee interior point
u, v, w = rng.random(3)
s = u + v + w
u, v, w = u / s, v / s, w / s
P = u * V1 + v * V2 + w * V3
x0, y0 = P

print(f"Initial point P = ({x0:.16f}, {y0:.16f})")
```

```python
# Randomly choose 1, 2, or 3 and move halfway toward that vertex
k = int(rng.integers(1, 4))
chosen_vertex = vertices[k - 1]
P_next = 0.5 * (P + chosen_vertex)

print(f"Random choice = {k}")
print(f"Next point = ({P_next[0]:.16f}, {P_next[1]:.16f})")

# Plot the triangle, P, and P_next
fig, ax = plt.subplots(figsize=(6, 6))
triangle = np.array([V1, V2, V3, V1])
ax.plot(triangle[:, 0], triangle[:, 1], 'k-', lw=2, label='Triangle edges')

ax.scatter(*V1, c='red', s=70)
ax.scatter(*V2, c='red', s=70)
ax.scatter(*V3, c='red', s=70)
ax.text(V1[0] - 0.05, V1[1] - 0.05, 'V1')
ax.text(V2[0] + 0.02, V2[1] - 0.05, 'V2')
ax.text(V3[0], V3[1] + 0.04, 'V3')

ax.scatter(*P, c='blue', s=60, label='P (initial)')
ax.scatter(*P_next, c='green', s=60, label='P_next (halfway step)')
ax.plot([P[0], chosen_vertex[0]], [P[1], chosen_vertex[1]], 'b--', alpha=0.6)
ax.plot([P[0], P_next[0]], [P[1], P_next[1]], 'g-', lw=2, alpha=0.8)

ax.set_aspect('equal', 'box')
ax.set_title('Sierpiński Chaos Game: First Step')
ax.legend(loc='upper right')
ax.grid(alpha=0.2)
plt.show()
```

### JC: After asking the agent to make a triangle and choosing to make a point and then give a condition for the next point. Given this specific random choice the next point is halfway between v3 and p initial. Next I should explore larger numbers of p_next.

```python
# Generate 500 more points using the last P_next as the current point each step
n_more = 500

current_point = P_next.copy()
generated_points = []
chosen_indices = []

for _ in range(n_more):
    k = int(rng.integers(1, 4))  # 1, 2, or 3
    chosen_vertex = vertices[k - 1]
    current_point = 0.5 * (current_point + chosen_vertex)
    generated_points.append(current_point.copy())
    chosen_indices.append(k)

generated_points = np.array(generated_points)
P_next = current_point  # keep latest point available for future cells

print(f"Generated {len(generated_points)} additional points.")
print(f"Final P_next = ({P_next[0]:.16f}, {P_next[1]:.16f})")

# Plot triangle and the 500 generated points
fig, ax = plt.subplots(figsize=(7, 7))
triangle = np.array([V1, V2, V3, V1])
ax.plot(triangle[:, 0], triangle[:, 1], 'k-', lw=2)

ax.scatter(generated_points[:, 0], generated_points[:, 1], s=6, c='purple', alpha=0.75, label='500 new points')
ax.scatter(*P, c='blue', s=30, label='Initial P')
ax.scatter(*generated_points[-1], c='orange', s=45, label='Last generated point')

ax.set_aspect('equal', 'box')
ax.set_title('Sierpiński Chaos Game: 500 Additional Points')
ax.legend(loc='upper right')
ax.grid(alpha=0.2)
plt.show()
```

### JC: To make it even crispier I should make an animation that zooms into the top vertex.

```python
from pathlib import Path
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import HTML, Image, display

# Use previously generated points
if "generated_points" not in globals() or len(generated_points) == 0:
    raise ValueError("Run a point-generation cell first so generated_points is available.")

points = generated_points
n_pts = len(points)

# Keep animation responsive if point count is huge
max_points_for_anim = 150_000
if n_pts > max_points_for_anim:
    idx = np.linspace(0, n_pts - 1, max_points_for_anim, dtype=int)
    points_to_plot = points[idx]
else:
    points_to_plot = points

# Triangle and zoom target (top vertex V3)
triangle = np.array([V1, V2, V3, V1])
vx, vy = float(V3[0]), float(V3[1])

x_full = (float(np.min(triangle[:, 0])) - 0.05, float(np.max(triangle[:, 0])) + 0.05)
y_full = (float(np.min(triangle[:, 1])) - 0.05, float(np.max(triangle[:, 1])) + 0.05)

# Final window size around top vertex
half_w_final = 0.03
half_h_final = 0.03
x_final = (vx - half_w_final, vx + half_w_final)
y_final = (vy - half_h_final, vy + half_h_final)

n_frames = 80
fps = 20

fig, ax = plt.subplots(figsize=(7, 7))

def lerp(a, b, t):
    return a + (b - a) * t

def update(frame):
    t = frame / (n_frames - 1)
    ax.clear()  # redraw from scratch so each frame is a fresh plot

    # Interpolate axis bounds toward the top vertex window
    x0 = lerp(x_full[0], x_final[0], t)
    x1 = lerp(x_full[1], x_final[1], t)
    y0 = lerp(y_full[0], y_final[0], t)
    y1 = lerp(y_full[1], y_final[1], t)

    ax.plot(triangle[:, 0], triangle[:, 1], color="black", lw=1.5, alpha=0.7)
    ax.scatter(points_to_plot[:, 0], points_to_plot[:, 1], s=0.2, c="purple", alpha=0.7)
    ax.scatter(vx, vy, c="red", s=30)

    ax.set_xlim(x0, x1)
    ax.set_ylim(y0, y1)
    ax.set_aspect("equal", "box")
    ax.set_title(f"Zoom to Top Vertex (frame {frame + 1}/{n_frames})")
    ax.grid(alpha=0.2)

    return []

ani = FuncAnimation(fig, update, frames=n_frames, interval=1000 / fps, blit=False)

gif_path = Path.cwd() / "sierpinski_top_vertex_zoom.gif"
ani.save(gif_path, writer=PillowWriter(fps=fps))
html_anim = ani.to_jshtml(fps=fps)
plt.close(fig)

display(HTML(html_anim))
display(Image(filename=str(gif_path)))
print(f"Saved GIF: {gif_path}")
```
