```python
#part a)
from numpy import array

#first lay out all the constants we are given and any variables we want
a1, a2, a3, a4 = 15.8, 18.3, 0.714, 23.2 
A, Z = 58, 28
#we can make an if statement to define a5
if A%2>0:
    a5 = 0
elif Z%2>0:
    a5 = -12
elif Z%2 == 0:
    a5 = 12
#now we can define our function B
B = a1*A - a2*(A)**(2/3) - a3*((Z)**2)/(A**(1/3)) - a4*((A-2*Z)**2/A) + a5/(A**(1/2))

#now we can print out B or B/A
print(np.round(B, decimals = 1), "MeV")
```

```python
#part b)
#from previous cell we can use previous code

BPN = B/A #BPN is Binding energy per nucleon
print(np.round(BPN, decimals = 3), "MeV/Nucleon")
```

```python
#part C) 
#using the same code we only need to modify A and Z and put it into an array and sorts out the most stable BPN
Z = 82
A = Z
r = [] #declare an empty list

#use for loop to append the array with BPN
for A in range(Z, 3*Z+1):
    B = a1*A - a2*(A)**(2/3) - a3*((Z)**2)/(A**(1/3)) - a4*((A-2*Z)**2/A) + a5/(A**(1/2))
    BPN = B/A
    r.append(BPN)
    MaxBPN = np.nanmax(r)
    print("Mass number A:", A, "with Binding Energy", BPN)
print(MaxBPN, "Largest Binding energy per nucleon which has an atomic mass of")


```
