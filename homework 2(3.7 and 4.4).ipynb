```python
#problem 3.7
import numpy as np
import matplotlib.pyplot as plt

N = 400
m = 10000
g = 2 #makes a space from -2, 2 with n spacing 
p = 2 #change the shape
x = np.linspace(-g, g, N)
y = np.linspace(-g, g, N)
#make an array to occupy the mandelbrot
mbro = np.zeros((N, N))
```

```python
#goes through every point in x and y and asks is it apart of the mandelbrot
for i in range(N):
    for j in range(N):
        c = x[j] + 1j*y[i] #computation for c that scans every x(j) and y(i) coordinate because x  is columns and y is rows and arrays fit mbro[row, column]
        z = 0
        for n in range(m):
            z = z**p + c 
            if abs(z) > 2:
                mbro[i,j] = 1
                break
                
```

```python
plt.imshow(mbro, extent=[-g, g, -g, g], cmap = 'gray_r')
plt.show()
```

```python
import numpy as np
import math

def f(x):
    return np.sqrt(1-x**2)

def trapz(f, a, b, N):
    h = (b - a) / N
    total = 0
    x = a
    for i in range(N):
        total += (h * (f(x) + f(x + h)) / 2)
        x += h
    return total
#n = 1000000
n = 10000
print(trapz(f, -1, 1, n))
print(np.pi / 2)
```
